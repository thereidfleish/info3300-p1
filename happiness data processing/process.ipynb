```python
import pandas as pd
```

```python
batch = pd.read_csv("last batch.csv").drop(["time", "sunrise (iso8601)", "sunset (iso8601)"], axis=1)
batch
```

```python
batch["daylight_duration (s)"] = batch["daylight_duration (s)"] / 3600
batch["sunshine_duration (s)"] = batch["sunshine_duration (s)"] / 3600
```

```python
gb = batch.groupby("location_id").mean().reset_index()
gb = gb.add_suffix('_mean')
gb
```

```python
# gb.to_csv("first_batch_means.csv")
gb.to_csv("last_batch_means.csv")
```
